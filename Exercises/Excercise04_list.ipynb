# 1. Dice rolls

```python
import random as rnd

dice_rolls = []

for i in range(10):
    dice_rolls.append(rnd.randint(1,6))

print(f"Dice rolls : {dice_rolls}")

dice_rolls.sort()
print(f"Dice rolls in ascending order: {dice_rolls}")

dice_rolls.sort(reverse=True)
print(f"Dice rolls in descending order: {dice_rolls}")

print(f'Max value: {dice_rolls[0]}')
print(f'Min value: {dice_rolls[-1]}')
```

# 3. Squares

```python
import matplotlib.pyplot as plt

numbers = [x for x in range(21)]
squares = [x**2 for x in range(21)]

plt.plot(range(21), squares, "o-")
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.grid()
```

# 4. Chessboard

```python
chess_letters = (f'ABCDEFGH')
chess_numbers = (f'12345678')
chess_list = [[letter+number for letter in chess_letters] for number in chess_numbers]

for row in chess_list:
    print(row)
```
