# 1. Dice rolls

```python
import random as rnd

dice_rolls = []

for i in range(10):
    dice_rolls.append(rnd.randint(1,6))

print(f"Dice rolls : {dice_rolls}")

dice_rolls.sort()
print(f"Dice rolls in ascending order: {dice_rolls}")

dice_rolls.sort(reverse=True)
print(f"Dice rolls in descending order: {dice_rolls}")

print(f'Max value: {dice_rolls[0]}')
print(f'Min value: {dice_rolls[-1]}')
```

# 3. Squares

```python
import matplotlib.pyplot as plt

numbers = [x for x in range(21)]
squares = [x**2 for x in range(21)]

plt.plot(range(21), squares, "o-")
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.grid()
```

# 5. Dice rolls convergence

```python
import random as rnd
import matplotlib.pyplot as plt

dice_rolls = [ ]
conv_list = [ ]
sixes = 0
stop = 1

for i in range(100):
    dice_rolls.append(rnd.randint(1,6))
    if dice_rolls[i] == 6:
        sixes += 1
print(f"Percentage on 100 rolls: {100*sixes/100:.0f}%")

dice_rolls = []

for i in range(6):
    sixes = 0
    stop *= 10

    for j in range(stop):
        dice_rolls.append(rnd.randint(1,6))
        if dice_rolls[j] == 6:
            sixes+=1
    conv_list.append(sixes/stop)

conv_list

plt.plot(conv_list, "o-")
plt.grid()
plt.show()
```
