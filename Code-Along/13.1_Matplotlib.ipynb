# Matplotlib
- MATLAB - style
- OOP approach

```python
# Setup
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5)
f = np.sin(x)

#print(x)
```

## MATLAB style interface

```python
plt.figure(dpi=100)
plt.plot(x,f)
plt.title("Sin X")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.figure(dpi=100, figsize=(10,4))
plt.subplot(1,2,1) # (rows, cols, panel number)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.subplot(1,2,2) 
plt.plot(x, np.cos(x))
plt.show()
```

# OOP approach

```python
fig = plt.figure(dpi=100)
# ax object can have a plot axes, labels, title,...
ax = plt.axes()
print(fig)

ax.plot(x, f)
ax.set(xlabel="X", ylabel="Y", title="$\sin{x}$", xlim=(0,5));
```

```python
fig, ax = plt.subplots(1,2, dpi=100, figsize=(10,4))
rng = np.random.RandomState(42)
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha=.5); # bins = antal staplar, alpha = hur transparant, ";" tar bort utskriften
ax[0].set(title="Histogram of normal distributed data", xlabel="X", ylabel="Y", );

bubble = ax[1].scatter(x=rng.randn(100), y=rng.randn(100), c = rng.randn(100), alpha=.7, cmap="cividis", s = 1000*rng.randn(100))
ax[1].set(title="Bubble Chart", xlabel="X", ylabel="Y")
fig.colorbar(bubble);

plt.savefig("beautiful_plots.png", facecolor="white")
```
